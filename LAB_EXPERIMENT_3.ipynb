```python
# Interpolation Search on Floating Point Numbers
# Dataset sizes: 10000, 50000, 100000


import random
import matplotlib.pyplot as plt
import pandas as pd



# ---------------------------------
# Interpolation Search for Floats
# ---------------------------------

def interpolation_search(arr, target):

    low = 0
    high = len(arr) - 1

    comparisons = 0


    while low <= high and arr[low] <= target <= arr[high]:

        comparisons += 1


        # Interpolation formula for floating values

        pos = low + int(
            ((target - arr[low]) * (high - low)) /
            (arr[high] - arr[low])
        )


        comparisons += 1


        if arr[pos] == target:

            return pos, comparisons



        comparisons += 1


        if arr[pos] < target:

            low = pos + 1


        else:

            high = pos - 1



    return -1, comparisons





# ---------------------------------
# Generate Floating Dataset
# ---------------------------------

def create_dataset(size):

    numbers = []


    for i in range(size):

        value = random.uniform(0.0, 1000.0)

        numbers.append(value)


    numbers.sort()


    return numbers





# ---------------------------------
# Main Program
# ---------------------------------


dataset_sizes = [
    10000,
    50000,
    100000
]


results = []



for size in dataset_sizes:


    # Create dataset

    data = create_dataset(size)



    # Select random value from dataset

    target = random.choice(data)



    # Search

    index, comparisons = interpolation_search(
        data,
        target
    )



    results.append([

        size,

        target,

        comparisons

    ])





# ---------------------------------
# Display Table
# ---------------------------------


df = pd.DataFrame(

    results,

    columns=[

        "Dataset Size",

        "Target Value",

        "Number of Comparisons"

    ]

)



print("\nInterpolation Search Analysis")

print("-----------------------------")


print(df)





# ---------------------------------
# Chart Analysis
# ---------------------------------


plt.figure(figsize=(8,5))



plt.plot(

    df["Dataset Size"],

    df["Number of Comparisons"],

    marker="o"

)



plt.xlabel("Dataset Size")


plt.ylabel("Number of Comparisons")


plt.title(
    "Interpolation Search Comparison Analysis"
)


plt.grid(True)


plt.show()
```
